```python
#feature [1 sim, 0 não]
#pelo longo
#perna curta
#faz au-au


porco1 = [0,1,0]
porco2 = [0,1,1]
porco3 = [1,1,0]

cachorro1 = [0,1,1]
cachorro2 = [1,0,1]
cachorro3 = [1,1,1]
#normalmente os dados se chamam x
treino_x = [porco1, porco2, porco3, cachorro1, cachorro2, cachorro3] #dados
# aqui chamamos os porcos de 1 e o cachorro de 0
#normalmente as classes que classificam o que é o que, se chama y
treino_y = [1,1,1,0,0,0] #classes
```

```python
from sklearn.svm import LinearSVC

modelo = LinearSVC()
modelo.fit(treino_x, treino_y) #entregando os dados dos animais e as classes para iniciar o treino da IA
```

```python
animal_misterioso = [0,1,0]
modelo.predict([animal_misterioso])
#1 = porco/ 0 = cachorro
# a IA acertou
```

```python
misterio1 = [1,1,1]
misterio2 = [1,1,0]
misterio3 = [0,1,1]
teste_x = [misterio1, misterio2, misterio3] #teste
previsoes = modelo.predict(teste_x)
```

```python
testes_y = [0,1,1] #Gabarito
```

```python
previsoes
```

```python
#gerar a % de acurácia na mão sem função pronta
corretos = (previsoes == testes_y).sum()
total = len(teste_x)
taxa_de_acerto = corretos/total *100
print(f"Acurária: {taxa_de_acerto:.2f}")
#conta para saber a % de acurária que a IA retorna
```

```python
##sda linha 9 abaixo nós estavamos fazendo um teste com a IA para ver o quanto ele aprendeu, fizemos animais
##misteriosos e não falamos quem é quem e o testes_classes é o gabarito mas ele não sabe disso e quando chamo previões ele da o resultado do que ele fala o que acha que é
##dessa maneira sabemos a acurácia das respostas retornadas pela IA
```

```python
##essa é para gerar pontuação de acurácia com função pronta
from sklearn.metrics import accuracy_score
taxa_de_acerto = accuracy_score(testes_y, previsoes) * 100
print(f"Acurária: {taxa_de_acerto:.2f}")
```
